```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
AM_G = pd.read_table("/Users/cbenson/Bodac/DE/DESeq2_AM_gene/Kallisto.gene.count.matrix.AM.wAnnot.matrix")
AM_I = pd.read_table("/Users/cbenson/Bodac/DE/DESeq2_AM_isoform/Kallisto.isoform.count.matrix.AM.wAnnot.matrix")
PM_G = pd.read_table("/Users/cbenson/Bodac/DE/DESeq2_PM_gene/Kallisto.gene.count.matrix.PM.wAnnot.matrix")
PM_I = pd.read_table("/Users/cbenson/Bodac/DE/DESeq2_PM_isoform/Kallisto.isoform.count.matrix.PM.wAnnot.matrix")
```

```python
AM_G.sort_values(by='padj')
AM_I.sort_values(by='padj')
PM_I.sort_values(by='padj')
PM_G.sort_values(by='padj')
```

```python
AI_pcut = AM_I[AM_I.padj < 0.05]
PI_pcut = PM_I[PM_I.padj < 0.05]
AG_pcut = AM_G[AM_G.padj < 0.05]
PG_pcut = PM_G[PM_G.padj < 0.05]
```

```python
AI_sort = AI_pcut.sort_values (by= 'log2FoldChange', ascending=False)
PI_sort = PI_pcut.sort_values (by= 'log2FoldChange', ascending=False)
PG_sort = PG_pcut.sort_values (by= 'log2FoldChange', ascending=False)
AG_sort = AG_pcut.sort_values (by= 'log2FoldChange', ascending=False)
```

```python
AI_up = AI_sort[AI_sort.log2FoldChange > 0]
AG_up = AG_sort[AG_sort.log2FoldChange > 0]
PI_up = PI_sort[PI_sort.log2FoldChange > 0]
PG_up = PG_sort[PG_sort.log2FoldChange > 0]
```

```python
AI2 = AI_pcut.sort_values (by= 'log2FoldChange', ascending=True)
PI2 = PI_pcut.sort_values (by= 'log2FoldChange', ascending=True)
AG2 = AG_pcut.sort_values (by= 'log2FoldChange', ascending=True)
PG2 = PG_pcut.sort_values (by= 'log2FoldChange', ascending=True)
```

```python
AI_down = AI2[AI2.log2FoldChange > 0]
AG_down = AG2[AG2.log2FoldChange > 0]
PI_down = PI2[PI2.log2FoldChange > 0]
PG_down = PG2[PG2.log2FoldChange > 0]
```

```python
AI_up.to_csv('Bodac_AM_iso_up.txt', sep='\t')
PI_up.to_csv('Bodac_PM_iso_up.txt', sep='\t')
AG_up.to_csv('Bodac_AM_gene_up.txt', sep='\t')
PG_up.to_csv('Bodac_PM_gene_up.txt', sep='\t')
```

```python
AI_down.to_csv('Bodac_AM_iso_down.txt', sep='\t')
PI_down.to_csv('Bodac_PM_iso_down.txt', sep='\t')
AG_down.to_csv('Bodac_AM_gene_down.txt', sep='\t')
PG_down.to_csv('Bodac_PM_gene_down.txt', sep='\t')
```
